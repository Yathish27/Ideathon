```python
#power geneartion using hydro project
x=float(input("Enter the inlet quantity in litres:"))
y=x*0.264
alt=float(input("Enter the Altitude difference of two points of lake for channelizing water into turbine in m:"))
Energy=y*3.79*alt*9.81
Power=Energy/(24*3600)
print("Power in kW generated by Turbine in Hydro Project=",Power,"kW")
```
